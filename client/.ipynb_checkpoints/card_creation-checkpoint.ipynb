```python
class card:
     def __init__(self, name, cost, move_f, move_ch, move_tm, level, a_iv, d_iv, hp_iv, tm_flag, shiny, attack_stg, def_stg, health, energy, bonus_shield, status):
        self.name = name
        self.cost = cost
        self.move_f = move_f
        self.move_ch = move_ch
        self.move_tm = move_tm
        self.level = level
        self.a_iv = a_iv
        self.d_iv = d_iv
        self.hp_iv = hp_iv
        self.tm_flag = tm_flag
        self.attack_stg = attack_stg
        self.def_stg = def_stg
        self.shiny = shiny
        self.health = health
        self.energy = energy
        self.bonus_shield = bonus_shield
        self.status = status    


def card_from_string(pvpokestring):
    l = pvpokestring.split(',')
    return(card(
    name = l[0],
    cost = l[8],
     move_f = l[1],
     move_ch = l[2],
     move_tm = l[3],
     level = l[4],
    a_iv = l[5], 
    d_iv = l[6], 
    hp_iv = l[7],
    tm_flag = 0,
    attack_stg = 0,
    def_stg = 0,
    shiny = 0,
    health = 0,
    energy = 0,
    bonus_shield= 0,
    status = 1
        
    )
          )
```
